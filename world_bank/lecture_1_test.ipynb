```python
def get_indicator(country_code, indicator_code):
    url = 'http://api.worldbank.org/v2/country/'
    url += country_code
    url += '/indicator/'
    url += indicator_code
    url += '?format=json'
    r = requests.get(url)
    try:
        data = r.json()
        for each in data[1]:
            if each['value'] is None:
                continue
            else:
                return each['value']
    except KeyError:
        return None
    except IndexError:
        return None
    except:
        return None
```

```python
def get_iso_code(country_name):
    try: 
        url = 'https://restcountries.eu/rest/v2/name/' + country_name 
        r = requests.get(url) 
        data = r.json() 
        return data[0]['alpha3Code']
    except KeyError:
        print('[ERROR] cannot get code for', country_name) 
        return None
get_iso_code('Thailand')
```

```python
countries = {}
for each in df['Country (according to CN Gov):']:
    if 'BRI' in each:
        each = each.split('(')[0].strip()
    iso = get_iso_code(each)
    if iso is not None:
        countries[each] = iso
countries['Thailiand'] = 'THA'
countries['The United Arab Emirates'] = 'ARE'
```

```python
print(countries)
```

```python
for i,v in df['Country (according to CN Gov):'].iteritems():
    country_name = v.split('(')[0].strip()
    country_code = countries[country_name]
    value = get_indicator(country_code,'GC.TAX.IMPT.ZS')
    if value is not None:
        df.at[i,'IMF WB\nCustoms and other import duties (% of tax revenue) 2016'] = value
    print(country_name,'done...')
    time.sleep(0.02)
df[['Country (according to CN Gov):','IMF WB\nCustoms and other import duties (% of tax revenue) 2016']]
```
